```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('fig1.jpg')
pnt_img1 = np.array([[446,72], [442,230], [383,102], [353,195]])

img2 = cv2.imread('fig2.jpg')
pnt_img2 = np.array([[216,66], [470,114], [259,157], [388,220]])

fig = plt.figure(figsize=(10,20))

fig.add_subplot(1, 2, 1)
plt.imshow(img1)

fig.add_subplot(1, 2, 2)
plt.imshow(img2)
```

```python
h, status = cv2.findHomography(pnt_img1, pnt_img2)
img_out = cv2.warpPerspective(img1, h, (img2.shape[1], img2.shape[0]))

plt.figure(figsize=(5,10))
plt.imshow(img_out)
plt.show()
```
